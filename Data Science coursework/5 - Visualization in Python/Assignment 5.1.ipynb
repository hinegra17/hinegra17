## Part 2.1: Some really nice plots

```python
sns.get_dataset_names()
```

```python
crashes = sns.load_dataset('car_crashes')
crashes
```

I noticed after making my first plots that if I reordered the columns in my dataframe, it would make my legend easier to interpret later on. Furthermore, I'm going to drop a couple columns to make melting the data I want into specific columns easier.

```python
crashes.columns
```

```python
crashes = crashes[['total','no_previous','not_distracted','speeding', 'alcohol']]
crashes
```

```python
#renaming columns for the legend later on
crashes = crashes.rename(columns = {"total" : "Fatal Crashes per 1b Miles",
                         "speeding" : "Drivers Who Were Speeding",
                         "alcohol" : "Drivers Who Were Alcohol-Impaired",
                         "not_distracted" : "Drivers Who Were Not Distracted",
                         "no_previous" : "Drivers With No Previous Accidents"})
crashes
```

```python
melt_perc = crashes.melt("Fatal Crashes per 1b Miles", value_name = "Percentage of Fatal Accidents", var_name = "Accident Description")
melt_perc
```

```python
sns.set_theme() #changing to seaborns default theme
sns.set_context("notebook", font_scale = 1.2)
plot = sns.jointplot(data = melt_perc, 
              x = "Fatal Crashes per 1b Miles", y = "Percentage of Fatal Accidents",
              hue = "Accident Description", height = 10, ratio = 3, palette = "bright", 
              marginal_kws = {"shade" : False})
```

My goal with this plot was to show any relationships between the types of car accidents (as represented by "Percentage of Fatal Accidents") and the total amount of crashes in a given state, As well as the distributions of the data. It's important to note that these categories are not mutually exclusive, i.e. just because a car accident occurred when a driver was speeding does not mean that that the driver was not distracted.

Things to notice in the data:
- The distribution of fatal crashes per 1 billion miles for all states is normally distributed.
- The distributions of fatal accidents where the drivers had either no previous accidents or they were not distracted are also normally distributed, but the distributions seem to be very spread out.
- The distributions of fatal accidents where drivers were speeding or alcohol impaired are more tightly distributed from 0-10%, and both distributions don’t seem to be perfectly gaussian. The Alcohol impaired drivers distribution has a few outliers around 10%, and the speeding drivers distribution looks more like a mountain than a curve.
- As to the relationships between the types of accidents and total fatal crashes in a state, there is one noteworthy takeaway. While higher percentages of all types of accidents look to be correlated with higher amounts of total crashes, accidents involving drivers with no previous accidents or undistracted drivers look to be more tightly correlated with total accidents than the other two types of accidents.



Design decisions:
-	I chose to do a joint plot combining scatterplots and density curves because I wanted to clearly visualize any relationship between the percentages of accidents and total crashes, as well as the distributions of those accident types.
-	I changed the theme and font scale to make it more visually appealing and cleaner. Furthermore, seaborn’s default theme allows for easier identification of the locations of datapoints due to the gridlines.
-	I feel my data to ink ratio is great, as there is barely any ink in the graphic not representing data. My first draft of the graphic had shading underneath the density plots, which ended up having too much overlapping shading and made interpreting and comparing the plots difficult. I changed this with `marginal_kws = {"shade" : False}`.
-	I changed the color of data with `palette = "bright"` to make differentiating the different parts of the data easier, as I thought the default palette did not distinguish the types of data adequately.
- Changing `height = 10` and `ratio = 3` was intended to give more room to the data such that it could be seen better by the viewer. Originally, there was too much data in small spaces, making it hard to figure out what was going on.
