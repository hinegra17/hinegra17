**What is the dataset about?**

This dataset revolves around housing prices. This information shows housing price indexes that measure single-family home values, based off data that was collected from all fifty states and over four-hundred cities nationwide. This is also including the United States territory of Puerto Rico. This data spans from 1991 through this year. It includes the states, cities, and all the possible housing price index changes throughout the years.

The HPI index from the Federal Housing and Finance Agency helps people determine if it is a good time to buy a house or property in certain areas of a given area. HPI is determined based on repeat transactions, involving conventional and conforming mortgages on single family properties. It is a weighted repeat sales index that measures average price changes in repeat sales or refinancing on the same properties. In short, higher HPI's mean that the prices of homes are higher. This dataset allows the comparison of different HPI’s throughout long periods of time, in this case we plan to be looking at the HPI’s from 2000-Present Day and organizing by state.

**How/from where did you acquire it?**

This dataset is from the website of the Federal Housing and Finance Agency. # Part 2: Coding and Analysis

Now let's take a look at all the different place names in our data set.

```python
newCentury["place_name"].unique()
```

This may be hard to filter, so lets look at the 'level' column instead

```python
newCentury["level"].value_counts()
```

We have data ranging from specific cities to country wide. To compare HPI's accross states, we're going to just narrow it down to state level data.

Looking at only state data:

```python
states = newCentury[newCentury["level"] == "State"].copy()
states
```

Now to make it so that we can compare all the data from all these states across time, we need to create a column in datetime format.

First we're going to create a column called "month" where all of the period observations (which are labeled 1-4 for the corresponding quarter the year it is identifying) are converted to the month in which the given quarter ends. For example, all of the periods labeled '2' will be converted to '6' (which means June), because the second quarter of the year ends at the end of June.

```python
states["month"] = states["period"] * 3
states
```

Now we will use the "Year" and the "month" columns to create a new datetime format column called "Date":

```python
#Making every quarter end at on the 30th because some months only have 30 days
states["Date"] = states["month"].astype(str) + "/30/" + states['yr'].astype(str)
states["Date"] = pd.to_datetime(states["Date"])
states["Date"]
```

```python
states.sort_values(by = ['Date'])
```

Now we will look at only data that has the HPI Type of 'Traditional' and HPI flavor "expanded-data" (reasoning for so given in *Part 1*)

```python
mask1 = states["hpi_type"] == "traditional"
mask2 = states["hpi_flavor"] == "expanded-data"

tradexpanded = states[mask1 & mask2].copy()
tradexpanded.sort_values(by = ["Date", "place_name"], inplace = True)
tradexpanded
```

```python
#moving date to the left just to make it easier to read
date = tradexpanded.pop("Date")
tradexpanded.insert(0, "Date", date)
tradexpanded
```

For the first part of our analysis, we're going to just look at the effects of COVID on HPI from 2017 to 2021. By doing so we will be able to compare a couple of years of data from what we consider as "normal times" to the pandemic years.

```python
#Making a function that returns a slice of the traditional dataframe with all of the data after the start of a given year
def afterYear(startyear):
    afterYeardf = tradexpanded[tradexpanded['yr'] >= startyear]
    return afterYeardf
```

```python
afterYear(2018) #for example this gives us all of the state level, traditionally typed data from 2018 onward
```

We're going to now make a line graph of all of the traditional data plotted against time for each state.

```python
afterYear(2018).plot.scatter(x = "Date", y ="index_nsa")
```

This is not very instructive at all, so lets divide the data by state.

```python
state_names = tradexpanded["place_name"].unique()
state_names
```

```python
for state in state_names:
    mask = afterYear(2018)["place_name"] == state
    afterYear(2018)[mask].plot.line(x = "Date", y = "index_nsa", title = state)
```

Main takaway: The trend that was occuring before COVID for most states of a steady increase in HPI (meaning an increase in the prices of existing homes) did not only continue, it exploded after the pandemic. For most states, the growth in the non-seasonally adjusted HPI went from a seemingly positive linear trend to almost an exponential trend after the summer of 2020. However, while many states experienced little to no dip in HPI around the start of the pandemic in March 2020, some states had a minor decrease.

Now we're going to look at which states had the biggest percent change in non-seasonally adjusted HPI from 2019 to the first quarter after the pandemic started in 2020 (June). To do so, we're going to compare the values of every state's NSA HPI from June 2019 to June 2021. We're going to use those dates because by comparing the same month we can control for seasonality, and we're going to use June because by the end of June 2020 is when the intitial effects of the pandemic were being felt on the economy (you know with all the lockdowns and stuff).

First we'll extract just June of 2019 and 2020.

```python
mask1 = tradexpanded["Date"] == "2019-06-30"
mask2 = tradexpanded["Date"] == "2020-06-30"
tradexpanded[mask1 | mask2]
```

Now we'll make a pivot_table to be able to easily compare the data for each state across the 2 years.

```python
pivot_table = tradexpanded[mask1 | mask2].pivot_table(values = "index_nsa", index = "place_name", columns = "Date")
pivot_table
```

Now we'll generate a column for the percent change in NSA HPI

```python
((pivot_table["2020-06-30"] / pivot_table["2019-06-30"]) * 100) - 100
```

```python
pivot_table["Percent Change from 2019-2020"] = ((pivot_table["2020-06-30"] / pivot_table["2019-06-30"]) * 100) - 100
pivot_table
```

Now we'll sort those values and look at the summary statistics.

```python
pivot_table["Percent Change from 2019-2020"].sort_values()
```

```python
pivot_table["Percent Change from 2019-2020"].describe()
```

Interestingly, every single state had an increase in HPI during the pandemic when compared to the previous year. The state with the highest percent increase in HPI was Idaho with an increase of 12.14%, the average increase was 6.16%, and the state with the lowest increase in HPI was North Dakota at 1.3%.

It's crazy that the prices of homes still continued to increase at a time that the economy was doing so poorly, but the news around the time kind of explained why. As people were stuck inside, people who could afford homes' income generally stayed the same or increased during the pandemic and lockdowns (because those in higher income brackets could work from home) while their spending went down. Combined with the fact that the demand for housing went up as people were spending all of their time inside, it makes sense that HPI increased from year to year for every state. More simply, when the people who could afford to buy houses now had the same or more money to do so and a new a desire to upgrade or change their living situation (being that they were stuck inside all day every day), it makes sense that housing prices went up.

Now that we've looked at the effects of COVID on NSA HPI from the years close to the pandemic, lets look the trend compared to 2008.

```python
for state in state_names:
    mask = afterYear(2001)["place_name"] == state
    afterYear(2001)[mask].plot.line(x = "Date", y = "index_nsa", title = state)
```

In the 2008 financial crisis, the HPI for most states plummeted (look at the drops for Illinois and Nevada in 2008!). This is likely due to the fact that the housing market collapsed. When comparing this to the finacial problems caused by COVID, its really interesting to see how the cause of a financial crisis leads to completely different results. COVID happened independent of a collapse in the housing market, so the financial effects of COVID didn't have a direct and severe negative effect on the housing market. However, the 2008 crisis was directly caused by a problem in the housing market, so in that case a drop in HPI is what caused a financial downturn, and therefore there were massive drops in HPI in the majority of states. Simply put, the COVID financial downturn had a small or even positive effect on housing prices, whereas the 2008 financial crisis had a direct negative effect on housing prices.|
